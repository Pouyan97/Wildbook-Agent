yt.videos returns title description and tags for the video as a list of tags. A new file will be made that has updated video descriptions and titles.

```python
def makeIDList():
    lineCount = 0
    Vids = []
    with open('ytIds.csv', 'r') as csvfile:
        csv_reader = csv.reader(csvfile, delimiter=',')
        for row in csv_reader:
            if(lineCount == 0):
                pass
            else:
                Vids.append(row[0])
            lineCount += 1
    return Vids

# to create a file with ID and Desc, Title, Tags.
def createCSV(VidsList):
    if VidsList == []:
        return
    for vid in VidsList:
        with open('ytDirectLocationID.csv', 'w+') as csvfile:
            
            filewriter = csv.writer(csvfile, delimiter=',',
                        quoting=csv.QUOTE_MINIMAL)
            
            filewriter.writerow(['Video ID','Title', 'Description', 'tags'])
        #find the video
            YTVideo = yt.videos(vid,True)
#             print(YTVideo)

        #no video found..
            if not YTVideo:
#                 print(vid)
                filewriter.writerow([vid,'','',''])
                continue
        #get the snippet
            snippet = YTVideo[0]['snippet']
#             print(snippet)
#             print()

            # if an item is not specified do so.
            if 'tags' not in snippet.values():
                snippet['tags'] = ''
            if 'title' not in snippet.values():
                snippet['title'] = ''
            if 'description' not in snippet.values():
                snippet['description'] = ''
        #insert into the csv file
            filewriter.writerow([vid, snippet['title'],snippet['description'],snippet['tags']])
        #make a dictionary
            dic = {}
            dic.append([vid, snippet['title'],snippet['description'],snippet['tags']])
            
```

```python
VidsList = []

VidsList = makeIDList()
# print(VidsList)
createCSV(VidsList)
```
